```python
import pandas as pd
import numpy as np
```

```python
# Cargamos, eliminamos duplicados y reseteamos el indice
movies_df = pd.read_csv("Data_Original/movies_dataset.csv")
movies_df.drop_duplicates(inplace=True)
movies_df = movies_df.reset_index(drop=True)
movies_df.head(3)
```

+ Eliminar las columnas que no serán utilizadas, **`video`**,**`imdb_id`**,**`adult`**,**`original_title`**,**`poster_path`** y **`homepage`**.

```python
movies_df.drop(['video'], axis=1,inplace=True)
movies_df.drop(['imdb_id'], axis=1,inplace=True)
movies_df.drop(['adult'], axis=1,inplace=True)
movies_df.drop(['original_title'], axis=1,inplace=True)
movies_df.drop(['poster_path'], axis=1,inplace=True)
movies_df.drop(['homepage'], axis=1,inplace=True)
```

```python
movies_df.head(3)
```

```python
movies_df.info()
```

Hacemos algunas observaciones sobre el dataset

```python
print("id's unicos: ",movies_df['id'].unique().shape[0])
print("Núemro de idiomas: ", movies_df["original_language"].unique().shape[0])
print("Cantidad de estatus: ", movies_df["status"].unique().shape[0])
```

Convertimos los tipos de datos 

```python
# Al tratar de convertit el tipo de dato nos percatamos de que contine caracteres no numéricos
movies_df[movies_df['budget'].str.len()>10]
```

```python
# ELiminaremos del data set los tres casos anteriores debido a la falta de información.
movies_df.drop([19725,29491,35575],inplace=True)
movies_df.reset_index(drop=True)
```

```python
# Ahora si procede la conversión del tipo de dato
movies_df['budget'] = movies_df['budget'].astype(float)
movies_df['popularity'] = movies_df['popularity'].astype(float)
```

```python
movies_df.info()
```

+ Los valores nulos de los campos **`revenue`**, **`budget`** deben ser rellenados por el número **`0`**.

```python
# Solo tenemos tres valor nulos en revenue
print(movies_df['revenue'].isnull().sum())
print(movies_df['budget'].isnull().sum())
```

```python
movies_df[movies_df['revenue'].isnull()]
```

```python
# Remplazamos los campos núlos por valor 0 en la columna 'revenue'
movies_df['revenue'].fillna(0,inplace=True)
```

```python
movies_df.info()
```

+ De haber fechas, deberán tener el formato **`AAAA-mm-dd`**, además deberán crear la columna **`release_year`** donde extraerán el año de la fecha de estreno.

```python
movies_df['release_date'].isnull().sum() # Tenemos 87 valores nulos
```

```python
# Eliminamos los nulos y verificamos 
movies_df.dropna(subset=['release_date'],inplace=True) 
movies_df['release_date'].isnull().sum()
```

```python
movies_df = movies_df.reset_index(drop=True) # Reseteamos el indice
```

```python
movies_df.info()
```

```python
# Convertimos a tipo de dato fecha la columna 'release_year'
movies_df['release_date'] = pd.to_datetime(movies_df['release_date'])
```

```python
# Agregamos una nueva columna con el año
movies_df['release_year'] = movies_df['release_date'].dt.year.astype(str)
```

```python
movies_df.info()
```

+ Crear la columna con el retorno de inversión, llamada **`return`** con los campos **`revenue`** y **`budget`**, dividiendo estas dos últimas **`revenue / budget`**, cuando no hay datos disponibles para calcularlo, deberá tomar el valor **`0`**.

```python
movies_df[movies_df['budget']==0].shape[0] # Numero de valores 0 de 'budget'
```

```python
# remplazamos el 0 por 1 por conveniencia 
movies_df['budget'].replace(0,1, inplace=True)
```

```python
movies_df[movies_df['budget']==0].shape[0]
```

```python
movies_df[movies_df['revenue']==0].shape[0] # Numero de valores 0 en 'revenue'
```

```python
movies_df['return'] = movies_df['revenue']/movies_df['budget']
```

```python
movies_df.info()
```
